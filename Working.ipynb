```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read in the combined.csv file
df = pd.read_csv('combined.csv')

#remove the unnamed column
df = df.drop('Unnamed: 0', axis=1)

#Count the number of 0s per column
zero_sum = df.isnull().sum()

#Count the number of rows in the dataframe
num_rows = df.shape[0]
print(num_rows)

#Sort the values in descending order
zero_sum = zero_sum.sort_values(ascending=True)

#Print the values
#print(zero_sum)

#Cut the list in half
zero_sum = zero_sum[0:37]

#Plot the values
zero_sum.plot(kind='barh', figsize=(10,10))

```
